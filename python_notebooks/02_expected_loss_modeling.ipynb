```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
loan_data = pd.read_csv("../data/clean_credit_portfolio.csv")
```

```python
#Assign Probability of Default (PD)
risk_weights = {

    'Current':0.01,

    'Bucket 1 (Low Risk)':0.02,

    'Bucket 2 (Medium Risk)':0.08,

    'Bucket 3 (High Risk)':0.25,

    'Default / Write-Off':1.00

}

loan_data['PD'] = (
    loan_data['DelinquencyTier']
             .map(risk_weights)
)
```

```python
#Loss Given Default
#Assume industry recovery rate
#Recovery Rate = 40%
loan_data['LGD'] = 0.60
```

```python
#Exposure at Default
loan_data['EAD'] = loan_data['OutstandingBalance']
```

```python
#Expected Loss
loan_data['Expected_Loss'] = (

    loan_data['PD']

    *

    loan_data['LGD']

    *

    loan_data['EAD']

)
```

```python
#Portfolio KPIs
#Total Expected Loss
loan_data['Expected_Loss'].sum()
```

```python
#Average Expected Loss
loan_data['Expected_Loss'].mean()
```

```python
#Maximum Exposure
loan_data['Expected_Loss'].max()
```

```python
#Portfolio Value
loan_data['loan_amount'].sum()
```

```python
#Total Outstanding Balance
loan_data['OutstandingBalance'].sum()
```

```python
#Expected Loss by Risk Bucket
loan_data.groupby(
    'DelinquencyTier'
)['Expected_Loss'].sum()
```

```python
#Expected Loss by Loan Status
loan_data.groupby(
    'loan_status'
)['Expected_Loss'].sum()
```

```python
#Top 10 Riskiest Loans
loan_data.sort_values(
    by='Expected_Loss',
    ascending=False
).head(10)
```

```python
#Export Final Dataset
loan_data.to_csv(
    "../data/final_credit_portfolio.csv",
    index=False
)
```
